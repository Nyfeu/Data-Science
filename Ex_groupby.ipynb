## **CASE Tips**

# Q1.

Considere a base de dados [`tips`](https://github.com/Rogerio-mack/IMT_Ciencia_de_Dados/raw/main/data/tips.csv). Considerando somente os finais de semana, quem tem média de participantes na mesa maior, fumantes ou não fumantes? **`[NoGroup | GroupBy]`**



```python
df = pd.read_csv('https://github.com/Rogerio-mack/IMT_Ciencia_de_Dados/raw/main/data/tips.csv')
df.head()
```

```python
df_tips = df.copy()

# Filtrando apenas finais de semana (Sat, Sun)
weekend_df = df_tips[df_tips['day'].isin(['Sat', 'Sun'])]

# Agrupando por fumante e calculando a média de 'size'
result = weekend_df.groupby('smoker')['size'].mean()

display(result)

if result['Yes'] > result['No']:
    print(f"\nFumantes têm uma média maior: {result['Yes']:.2f}")
else:
    print(f"\nNão fumantes têm uma média maior: {result['No']:.2f}")
```

## **CASE Insurance**

Acesse a base de dados [`insurance`](https://github.com/Rogerio-mack/IMT_Ciencia_de_Dados/raw/main/data/insurance.csv) para as questões a seguir.



```python
df = pd.read_csv('https://github.com/Rogerio-mack/IMT_Ciencia_de_Dados/raw/main/data/insurance.csv')
df.head()
```

# Q2.

Qual percentual de homens e mulheres assegurados na base? **`[NoGroup | GroupBy]`**

```python
# Calculando o percentual de homens e mulheres usando groupby
sex_counts = df.groupby('sex').size() / len(df) * 100

print("Percentual de assegurados por sexo (usando groupby):")
display(sex_counts.map('{:.2f}%'.format))
```

# Q3.

Qual o total do valor de seguro pago por homens e mulheres segurados na base? **`[NoGroup | GroupBy]`**

```python
total_value = df.groupby('sex')['charges'].sum()

print("Total do valor de seguro pago por homens e mulheres segurados na base:")
display(total_value.map('${:,.2f}'.format))
```

# Q4.

Qual o percentual de valor de seguros pagos por homens e mulheres segurados na base? **`[NoGroup | GroupBy]`**

```python
# Somatória de toda a receita
sum_total = df['charges'].sum()

# Calculando o percentual com base em total value por gênero
percentual_value = df.groupby('sex')['charges'].sum() / sum_total * 100

print("Percentual de valor de seguros pagos por homens e mulheres segurados na base:")
display(percentual_value.map('{:.2f}%'.format))
```

# Q5.

Não fumantes apresentam em média mais filhos? **`[NoGroup | GroupBy]`**

```python
children_mean = df.groupby('smoker')['children'].mean()

display(children_mean)
```

# Q6.

Qual a média e maior de idade dentre os segurados homem, das regiões northwest ou northeast? **`[NoGroup | GroupBy]`**

```python
# Calculando a média e a maior idade agrupado por sexo e região
result = df.groupby(['sex', 'region'])['age'].agg(['mean', 'max'])
filtered_df = result.loc[('male', ['northwest', 'northeast']), :]

print("Média e maior idade de homens nas regiões northwest e northeast:")
display(filtered_df)
```
